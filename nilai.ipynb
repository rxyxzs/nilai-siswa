```python
import pandas as pd 
import numpy as numpy
import matplotlib.pyplot as plt 

data = pd.read_csv("nilai_siswa.csv")
print(data.head())
# rata rata nilai tiap siswa
print("Rata-rata : ", data['Nilai'].mean())

rata_per_siswa = data.groupby('Nama')['Nilai'].mean()
print(rata_per_siswa)

#Hitung rata-rata per mata pelajaran
rata = data.groupby('Matpel')['Nilai'].mean()

#Tampilkan grafik
rata.plot(kind='bar', color='skyblue')
plt.title('Rata-rata Nilai per Mata Pelajaran')
plt.ylabel('Rata-rata Nilai')
plt.show()
```
